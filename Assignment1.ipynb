The top ten variables for prediction are: 

1. income amount total 
1. credit amount
1. goods price amount
1. annuity amount
1. credit sum amount
1. credit sum debt amount
1. balance amount
1. receivable amount
1. credit limit actual amount
1. down payment amount

The correlation values as seen below where used to determined the choices for the variables. Also, the use of the describe function from pandas was used to look at the mean, medium, and standard deviation. 

Other things considered when making the choice was for the income amount total because it should be necessary high to cover the annuity amount of the loan application and the balance amount on the credit card balance. Also, checking for the credit sum and credit sum debt amount could give an indication of the financial stability of the person. To mitigate the risk of the loan the goods price seems important for the decision and the receivalble amount last month to make sure the debt is getting paid. In addition, the down payment on the last loan is something good to check and the application amount from the last loan to see if the current request is more or less than what has been borrowed before.


```python
arr = {
    'target': ap['TARGET'],
    'key_id': ap['SK_ID_CURR'],
    'amount_income_total': ap['AMT_INCOME_TOTAL'],
    'amount_credit': ap['AMT_CREDIT'],
    'amount_goods_price': ap['AMT_GOODS_PRICE'],
    'amount_annuity':  ap['AMT_ANNUITY'],
    'amount_credit_sum': b['AMT_CREDIT_SUM'],
    'amount_credit_sum_debt': b['AMT_CREDIT_SUM_DEBT'],
    'amout_balance': ccb['AMT_BALANCE'],
    'amout_receivable': ccb['AMT_RECIVABLE'],
    'amount_application': pa['AMT_APPLICATION'],
    'amount_down_payment': pa['AMT_DOWN_PAYMENT']
    }
df = pd.DataFrame(arr)
corr = df.corr()
corr
```

```python
fig = plt.figure(figsize=(8,8), facecolor='white')
plt.matshow(corr, cmap='RdBu', fignum=fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr.columns)), corr.columns)
```
